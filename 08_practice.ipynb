```python
# common imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Use t-SNE to reduce the MNIST dataset down to two dimensions and plot the result using Matplotlib. You can use a scatterplot using 10 different colors to represent each image's target class.

```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original', data_home="datasets/mnist")
```

```python
np.random.seed(42)
idx = np.random.permutation(60000)[:10000]

X = mnist["data"][idx]
y = mnist["target"][idx]
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)
X_reduced = tsne.fit_transform(X)
```

```python
fig = plt.figure(figsize=(13,10))
plt.scatter(X_reduced[:,0], X_reduced[:,1],  edgecolors='black', c=y, cmap="jet")
plt.axis("off")
plt.colorbar()
plt.show()
```

```python
# plot digit 2, 3 and 5
fig = plt.figure(figsize=(10,10))

for digit, color in zip([2,3,5], ["blue", "red", "green"]):
    plt.scatter(X_reduced[y==digit, 0], X_reduced[y==digit, 1], c=color, edgecolors="black")
plt.axis("off")
plt.show()
```

```python
# running tsne on the three digits
mask = (y==2)|(y==3)|(y==5)
X_subset = X[mask]
y_subset = y[mask]
X_subset_reduced = TSNE(n_components=2).fit_transform(X_subset)
```

```python
fig = plt.figure(figsize=(11,7))
plt.scatter(X_subset_reduced[:,0], X_subset_reduced[:,1], c=y_subset, cmap="jet", edgecolors="black")
plt.colorbar()
plt.axis("off")
plt.show()
```
